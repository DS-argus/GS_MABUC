```python
from IPython.display import display, Math, Image
from identification import myID, mygID

from npsem.idty.gid import gID
from npsem.model import CD
```

#### Examples in [Tian, Pearl 2003]

```python
# Figure 1: Front-door case
# display(Image(filename='Images/03_Figure1.png'))

G = CD({'X', 'Z', 'Y'}, 
                    [('X', 'Z'), ('Z', 'Y')],
                    [('X', 'Y', 'U_XY')])

G.draw(pos={'X':(0,0),'Z':(1,0),'Y':(2,0)});

P = gID(G, {'Y'}, {'X'}, Zss=[set()])
print(f"Answer")
display(Math(str(P)))

P1 = myID(Y={'Y'}, X={'X'}, G=G)
print(f"myID")
display(Math(P1.printLatex()))

P2 = mygID(Y=set(["Y"]), X=set(["X"]), Z=frozenset([frozenset()]), G=G)
print(f"mygID")
display(Math(P2.printLatex()))
```

#### Examples in [Shpitser, Pearl 2006]

```python
# Figure 1.(a)
# display(Image(filename='Images/06_Figure1a.png'))

G = CD({'W1', 'W2', 'X', 'Y1', 'Y2'}, 
                    [('W1', 'X'), ('X', 'Y1'), ('W2', 'Y2')],
                    [('W1', 'W2', 'U_W1W2'), ('W1', 'Y2', 'U_W1Y2'), ('W2', 'X', 'U_W2X'), ('W1', 'Y1', 'U_W1Y1')])

G.draw(pos={'X':(1,1), 'W1':(0,1), 'W2':(0.5,0), 'Y1':(2,1), 'Y2':(1.5,0)});

P = gID(G, {'Y1', 'Y2'}, {'X'}, Zss=[set()])
print(f"Answer")
display(Math(str(P)))

P = myID(Y={'Y1', 'Y2'}, X={'X'}, G=G)
print(f"myID")
display(Math(P.printLatex()))

P2 = mygID(Y={'Y1', 'Y2'}, X={'X'}, Z=[set()], G=G)
print(f"mygID")
display(Math(P2.printLatex()))
```

```python
# Figure 1.(b)
# display(Image(filename='Images/06_Figure1b.png'))

G = CD({'W1', 'W2', 'X', 'Y1', 'Y2'}, 
                    [('W1', 'W2'), ('W1', 'X'), ('X', 'Y1'), ('W2', 'Y2')],
                    [('W1', 'W2', 'U_W1W2'), ('W1', 'Y2', 'U_W1Y2'), ('W2', 'X', 'U_W2X'), ('W1', 'Y1', 'U_W1Y1')])

G.draw(pos={'X':(1,1), 'W1':(0,1), 'W2':(0.5,0), 'Y1':(2,1), 'Y2':(1.5,0)});

try:
    P = gID(G, {'Y1', 'Y2'}, {'X'}, Zss=[set()])
    print(f"Answer")
    display(Math(str(P)))
except:
    print("gID Failed")

try:
    P1 = myID(Y={'Y1', 'Y2'}, X={'X'}, G=G)
except:
    print("myID Failed")

try:
    P2 = mygID(Y={'Y1', 'Y2'}, X={'X'}, Z=[set()], G=G)
except:
    print("mygID Failed")
```

#### Example in [Lee, Correa, Elias 2019]

```python
# Figure 1.(a)
# display(Image(filename='Images/19_Figure1a.png'))

G = CD({'X1', 'X2', 'W', 'Y'}, 
                    [('X1', 'W'), ('X2', 'Y'), ('W', 'Y')],
                    [('X1', 'W', 'U_X1W'), ('X1', 'X2', 'U_X1X2'), ('X2', 'Y', 'U_X2Y')])

G.draw(pos={'X1':(0,1), 'X2':(1,0.5), 'W':(0,0), 'Y':(1,-0.5)});

P = gID(G, {'Y'}, {'X1', 'X2'}, Zss=[{'X1'}, {'X2'}])
print(f"Answer")
display(Math(str(P)))

P1 = mygID(Y={'Y'}, X={'X1', 'X2'}, Z=[{'X1'}, {'X2'}], G=G)
print("mygID")
display(Math(P1.printLatex()))
```
